# Discrete Fourier Transform

Q1. Compute the 8-point DFT of the following sequences. Plot the magnitude and phase spectrum. Observe the symmetry properties.
<ol>
  <li>x[n] = {1,1,0,0,0,0,0,1}</li>
  <li>x[n] = {0,1,1,0,0,0,-1,-1}</li>
  <li>x[n] = {1,1,1,1,1,1,1,1}</li>
</ol>

https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.fft.fft.html#numpy.fft.fft

```python
import numpy as np
import matplotlib.pyplot as plt

#1
k = range(8)
a = np.array([1,1,0,0,0,0,0,1])
A = np.fft.fft(a,8)

print(A)

plt.stem(k,abs(A))
plt.title("magnitude spectrum")
plt.show()

angles = np.angle(A)
plt.stem(k, angles, 'g')
plt.title("phase spectrum")
plt.show()

# the magnitude spectrum is symmetric about n = 4
```

```python
#2
k = range(8)
a = np.array([0,1,1,0,0,0,-1,-1])
A = np.fft.fft(a,8)

print(A)

plt.stem(k,abs(A))
plt.title("magnitude spectrum")
plt.show()

angles = np.angle(A)
plt.stem(k, angles, 'g')
plt.title("phase spectrum")
plt.show()

#the magnitude spectrum is symmetric around n = 4 (the pattern from 0 to 4 repeats)
#the phase spectrum is anti-symmetric around n = 4
```

```python
k = range(8)
a = np.array([1,1,1,1,1,1,1,1])
A = np.fft.fft(a,8)

print(A)

plt.stem(k,abs(A))
plt.title("magnitude spectrum")
plt.show()

angles = np.angle(A)
plt.stem(k, angles, 'g')
plt.title("phase spectrum")
plt.show()

#the magnitude spectrum is zero everywhere but at n = 0 the value is 8
```

Q2. Generate the finite duration sequence $x[n] = 0.5(1-cos({\pi}n/20)), ~~0{\le}n{\lt}20$.
<ol>
  <li>Append the sequence with sufficient number of zeros to compute the DFT of the sequence with length N= 32 and 256. Plot the DFT and comment on the effects of zero padding</li>
  <li>Insert 16 zeros in the beginning of the sequence and repeat.</li>
</ol>


```python
n = np.arange(0,20)
x = 0.5*(1 - np.cos(np.pi*n/20))
x_ = np.append(x,np.zeros(12))  # padded with zeros for N= 32 
x1_ = np.append(x,np.zeros(236)) # padded with zeros for N= 256 
 

N_ = np.arange(0,32)
X_ = np.fft.fft(x_,32)
plt.stem(N_,abs(X_))
plt.title("magnitude spectrum for N = 32")
plt.show()
print(abs(X_))

angles = np.angle(X_)
plt.stem(N_, angles, 'g')
plt.title("phase spectrum for N = 32")
plt.show()

N1_ = np.arange(0,256)
X1_ = np.fft.fft(x1_,256)
plt.figure(figsize = (15,4))
plt.stem(N1_,abs(X1_))
plt.title("magnitude spectrum for N = 256")
xmin,xmax = plt.xlim()
plt.xlim(xmin*0.2,xmax)
plt.show()
print(abs(X1_))

angles1 = np.angle(X1_)
plt.figure(figsize = (15,4))
plt.stem(N1_, angles1, 'g')
plt.title("phase spectrum for N = 256")
xmin,xmax = plt.xlim()
plt.xlim(xmin*0.2,xmax)
plt.show()

#zero padding gives high density spectrum , improve frequency resolution ,interpolate between the original points of spectrum
#
```

```python
#2  INSERTED 16 ZEROS AT THE BEGINNING AND DONE  
N = np.arange(0,32)
n = np.arange(0,20)
x = np.zeros(16)  #added 16 zeros 
x = np.append(x,0.5*(1 - np.cos(np.pi*n/20)))


X = np.fft.fft(x,32)
plt.stem(N,abs(X))
plt.title("magnitude spectrum for N = 32 with 16 zeros")
plt.show()
print(abs(X))

angles = np.angle(X)
plt.stem(N, angles, 'g')
plt.title("phase spectrum for N = 32 with 16 zeros")
plt.show()

N = np.arange(0,256)
x1 = np.append(x,np.zeros(220))
X1 = np.fft.fft(x1,256)
plt.figure(figsize = (15,4))
plt.stem(N,abs(X1))
plt.title("magnitude spectrum for N = 256 with 16 zeros")
xmin,xmax = plt.xlim()
plt.xlim(xmin*0.2,xmax)
plt.show()
print(abs(X1))

angles = np.angle(X1)
plt.figure(figsize = (15,4))
plt.stem(N, angles, 'g')
plt.title("phase spectrum for N = 256 with 16 zeros")
xmin,xmax = plt.xlim()
plt.xlim(xmin*0.2,xmax)
plt.show()
```

Q4. Estimation of signal corrupted by noise: Generate the signal $s[n] = cos({\pi}5n/32) +cos({\pi}21n/64) $ for 256 samples. Generate a noisy version of the signal x[n]  by adding white gaussian noise e[n] with variance 0.3 to the original signal s[n]. Compute the energy in the error signal $\sum_{n}^{}(x[n] - s[n])^2$ 
<ul>
  <li>Plot the 256 point DFTs $S[k]$ and $X[k]$. Find the set $S_{nz}$ of the values of $k$ for which $S[k]$ is non-zero.</li>
  <li>Find $\hat{X}[k] = X[k]$ if  $k~ \in \mathcal ~{S_{nz}}$ and zero otherwise. Find the IDFT $\hat{x}[n]$ from $\hat{X}[k]$ to recover the original signal. Plot $\hat{x}[n]$. Find the error signal energy after recovery $\sum_{n}^{}(\hat{x}[n] - s[n])^2$</li>
</ul>

```python
n = np.arange(0,256)
s = np.cos(np.pi*5*n/32) + np.cos(np.pi*21*n/64)

e_noise = np.random.normal(0,np.power(0.3,0.5),256)
x = s + e_noise            #noisy version signal x[n]
 

#energy of signal
error_signal1 = np.sum(np.power(x - s,2))
print(error_signal1)          #energy in the error signal 

#plot DFT for S[k]
S = np.fft.fft(s,256)
plt.figure(figsize = (10,5))
plt.stem(n,abs(S))
plt.title("magnitude spectrum of 256 S[k]")
plt.show()
# print(abs(S))

angles = np.angle(S)
plt.figure(figsize = (15,4))
plt.stem(n, angles, 'g')
plt.title("Phase spectrum of 256 S[k]")
xmin,xmax = plt.xlim()
plt.xlim(xmin*0.2,xmax)
plt.show()

#plot DFT for X[k]
X = np.fft.fft(x,256)
plt.figure(figsize = (15,4))
plt.stem(n,abs(X))
plt.title("magnitude spectrum of 256 X[k]")
xmin,xmax = plt.xlim()
plt.xlim(xmin*0.2,xmax)
plt.show()

angles = np.angle(X)
plt.figure(figsize = (15,4))
plt.stem(n, angles, 'g')
plt.title("Phase spectrum of 256 X[k]")
xmin,xmax = plt.xlim()
plt.xlim(xmin*0.2,xmax)
plt.show()



#4) #2)   finding X cap k , Snz is the set of values of k where S[k] != 0
Snz = [20, 42, 214, 236]
Xk = (np.zeros(256))
Xk = Xk.astype(complex)
Xk[20] = X[20]
Xk[42] = X[42]
Xk[214] = X[214]
Xk[236] = X[236]

xk = np.fft.ifft(Xk,256)   #IDFT 
plt.figure(figsize = (15,4))
plt.stem(n,xk)
plt.title("x tilda [n] taken IDFT of X[k]")
xmin,xmax = plt.xlim()
plt.xlim(xmin*0.2,xmax)
plt.show()


error_signal2 = np.sum(np.power(xk - s,2))
print(abs(error_signal2))   #error signal energy after recovery
```

Q5. Generate the following signals $x_1[n] = cos({\pi}n/15),~~x_2[n] = cos(3{\pi}n/15)$ for n = 0 to 31. Plot the magnitude of 64 point DFT of the two signals $X_1[k]$ and $X_2[k]$ (with zero padding). Now concatenate the two signals to form a 64 length sequence $x[n]$. Plot the 64 point DFT magnitude $X[k]$ of $x[n]$. Also find the sum of two signals and plot the 64 point DFT of the result. (Based on the results, comment on the limitation of DFT.)

```python
N = np.arange(0,64)
n = np.arange(0,32)
x1 = np.cos(np.pi*n/15)
y1 = np.append(x1,np.zeros(32))
x2 = np.cos(3*np.pi*n/15)
y2 = np.append(x2,np.zeros(32))

X1 = np.fft.fft(y1,64)
plt.stem(N,abs(X1))
plt.title("Magnitude spectrum for X1[k]")
plt.show()

angles = np.angle(X1)
plt.stem(N,angles,'g')
plt.title("Phase spectrum for X1[k]")
plt.show()

X2 = np.fft.fft(y2,64)
plt.stem(N,abs(X2))
plt.title("Magnitude spectrum for X2[k]")
plt.show()

angles = np.angle(X2)
plt.stem(N,angles,'g')
plt.title("Phase spectrum for X2[k]")
plt.show()
```

```python
x = np.concatenate((x1,x2))
X = np.fft.fft(x,64)
plt.stem(N,abs(X))
plt.title("Magnitude spectrum for concatenate X1[k] and X2[k]")
plt.show()

angles = np.angle(X)
plt.stem(N,angles,'g')
plt.title("Phase spectrum for concatenate X1[k] and X2[k]")
plt.show()
```

```python
z = np.vstack([x1,x2])
z = np.sum([z[0],z[1]], axis=0)
z = np.append(z,np.zeros(32))

Z = np.fft.fft(z,64)
plt.stem(N,abs(Z))
plt.title("Magnitude spectrum for summation X1[k] and X2[k]")
plt.show()

angles = np.angle(Z)
plt.stem(N,angles,'g')
plt.title("Phase spectrum for sum X1[k] and X2[k]")
plt.show()
```
